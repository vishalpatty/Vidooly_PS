```python
#Libraries
import pandas as pd
import numpy as np
import re
from sklearn.ensemble import RandomForestRegressor
```

```python
raw = pd.read_csv('/home/deven/Desktop/Vidooly/train_data.csv')
raw.head()
```

```python
rawtest = pd.read_csv('/home/deven/Desktop/Vidooly/test_data.csv')
rawtest.head()
```

```python
#Converting Dates to Quarters
raw['qtr'] = pd.to_datetime(raw.published).dt.quarter
raw = raw.drop('published', axis = 1)
print raw.head()

rawtest['qtr'] = pd.to_datetime(rawtest.published).dt.quarter
rawtest = rawtest.drop('published', axis = 1)
print rawtest.head()
```

```python
z = raw.duration
new_dur = []
for each in z:
    num = re.findall(r'\d+', each)
    if len(num)==1:
        new_dur.append(int(num[0]))
    elif len(num)==2:
        new_dur.append((int(num[0])*60)+(int(num[1])))
    elif len(num)==3:
        new_dur.append(((int(num[0]))*3600)+(int(num[1])*60)+int(num[2]))
y = rawtest.duration
new_dur_test = []
for each in y:
    num = re.findall(r'\d+', each)
    if len(num)==1:
        new_dur_test.append(int(num[0]))
    elif len(num)==2:
        new_dur_test.append((int(num[0])*60)+(int(num[1])))
    elif len(num)==3:
        new_dur_test.append(((int(num[0]))*3600)+(int(num[1])*60)+int(num[2]))
```

```python
raw['duration_sec'] = new_dur
rawtest['duration_sec'] = new_dur_test
raw = raw.drop('duration', axis = 1)
rawtest = rawtest.drop('duration', axis = 1)
print raw.head()
print rawtest.head()
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
a = raw.category
b = rawtest.category
a_lab = LabelEncoder().fit_transform(a)
b_lab = LabelEncoder().fit_transform(b)
raw = raw.drop('category', axis = 1)
rawtest = rawtest.drop('category', axis = 1)
raw['category'] = a_lab
rawtest['category'] = b_lab
print raw.head()
print rawtest.head()
```

```python
s1 = raw.views
s2 = raw.likes
s3 = raw.dislikes
s4 = raw.comment
raw.views = pd.to_numeric(s1)
raw.likes = pd.to_numeric(s2)
raw.dislikes = pd.to_numeric(s3)
raw.comment = pd.to_numeric(s4)
```

```python
sw1 = rawtest.views
sw2 = rawtest.likes
sw3 = rawtest.dislikes
sw4 = rawtest.comment
rawtest.views = pd.to_numeric(sw1)
rawtest.likes = pd.to_numeric(sw2)
rawtest.dislikes = pd.to_numeric(sw3)
rawtest.comment = pd.to_numeric(sw4)
```

```python
test_id = rawtest.vidid
raw = raw.drop('vidid', axis = 1)
rawtest = rawtest.drop('vidid', axis = 1)
target = raw.adview
predictors = raw.drop(["adview"], axis =1)
```

```python
#Random Forest

rf_model = RandomForestRegressor(n_estimators=1000, random_state=42)
rf_model.fit(predictors, target)
rf_pred = rf_model.predict(rawtest)
print np.ceil(rf_pred)
```

```python
for i in range(len(rf_pred)):
    if rf_pred[i]<=0:
        rf_pred[i]=0
pred = np.ceil(rf_pred)
```

```python
d = {'vid_id': test_id}
sub = pd.DataFrame(data=d)
sub['ad_view'] = pred
sub.ad_view = sub.ad_view.astype(int)
```
